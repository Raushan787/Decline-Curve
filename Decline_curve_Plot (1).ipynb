```python
#Importing Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
# Data
t=np.linspace(1,1500)   #day
dates=pd.date_range(start="1/10/2000", end="1/10/23",periods=len(t))
date=dates.year
q_i=5000 #bbl/days
d=.005 # per day
t
```

Exponential Decline

```python
#Calculation of flowrate for exponential Decline
b=0
q_e=q_i*np.exp(-d*t)
q_e
```

```python
#Creating dataframe
df=pd.DataFrame({"t":t,"q":q_e},index=dates)
df["year"]=date
df.head()
```

```python
#Plotting the data
fig=px.line(data_frame=df,x="t",y="q",title="Flowrate vs Year")
fig.update_layout(
xaxis_title="Time (day)",
yaxis_title="Flowrate (bbl/day)",
template="plotly_dark")
fig.update_traces(line=dict(width=5,color='red'))
fig.show()
```

Harmonic Decline

```python
#Calculating the flowrate for harmonic decline
b=1
q_h=q_i/(1+d*t)
q_h
```

```python
#Creating datframe
df=pd.DataFrame({"t":t,"q":q_h},index=dates)
df["year"]=date
df.head()
```

```python
#Plotting the data

fig=px.line(data_frame=df,x="t",y="q",title="Flowrate vs Year")
fig.update_layout(
xaxis_title="Time (day)",
yaxis_title="Flowrate (bbl/day)",
template="plotly_white")
fig.update_traces(line=dict(width=5,color='green'))
fig.show()
```

Hyperbolic Decline

0<b<1

```python
#Take b as 0.3
b=0.3
```

```python
#Calculating flow rate for hyperbolic decline

q_hb=q_i/np.power((b*d*t+1),(1/b))
q_hb
```

```python
#Creating Dataframe
df=pd.DataFrame({"t":t,"q":q_hb},index=dates)
df["year"]=date
df.head()
```

```python
#Plotting the data
fig=px.line(data_frame=df,x="t",y="q",title="Flowrate vs Year")
fig.update_layout(
xaxis_title="Time (day)",
yaxis_title="Flowrate (bbl/day)",
template="seaborn")
fig.update_traces(line=dict(width=5,color='blue'))
fig.show()
```

```python
#Plotting data
plt.style.use("classic")
fig,ax=plt.subplots(nrows=3,ncols=1)
fig.set_figwidth(20)
fig.set_figheight(20)

#First Curve--> Exponential Curve
ax[0].plot(t,q_e,color="red",lw=6)
ax[0].set_title("Exponential Decline Curve")
ax[0].set_xlabel("Time (day)")
ax[0].set_ylabel("Q (bbl/day)")
ax[0].grid()
ax[0].legend()


#Second Curve-->Harmonic Curve
ax[1].plot(t,q_h,color="green",lw=6)
ax[1].set_title("Harmonic Decline Curve")
ax[1].set_xlabel("Time (day)")
ax[1].set_ylabel("Q (bbl/day)")
ax[1].grid()
ax[1].legend()


#Third Curve-->Hyperbolic Curve
ax[2].plot(t,q_hb,color="blue",lw=6)
ax[2].set_title("Hyperbolic Decline Curve")
ax[2].set_xlabel("Time (day)")
ax[2].set_ylabel("Q (bbl/day)")
ax[2].grid()
ax[2].legend()
```

```python
#Plot on same graph
plt.style.use("classic")
plt.figure(figsize=(20,10))
plt.plot(t,q_e,color="red",label="Exponential Decline Curve",lw=5)
plt.plot(t,q_h,color="green",label="Harmonic Decline Curve",lw=5)
plt.plot(t,q_hb,"b",label="Hyperbolic Decline Curve",lw=5)
plt.title("Decline Curve")
plt.xlabel("Time (day)")
plt.ylabel("Q (bbl/day)")
plt.legend()
plt.grid()
plt.legend()
```
