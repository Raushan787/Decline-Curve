```python
#Importing the Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

```

```python
# Lets assume required data
t=np.arange(1,1000)   #day
q_i=5000 #bbl/days
d=.005 # per day
```

**Exponential Decline**

The production rate of a well declines at a constant percentage over time.

```python
#Calculation of flowrate for exponential Decline
b=0
q_e=q_i*np.exp(-d*t)

#Cumulative production
q_e_cum=(q_i-q_e)/d
```

```python
#Creating dataframe
df=pd.DataFrame({"time":t,"q":q_e,"q_cum":q_e_cum})
df.head()
```

```python
#Plot
plt.figure(figsize=(15,5))
plt.plot(df["time"],df["q"],label="Flowrate vs time",lw=4)
plt.title("Plot for Exponential Decline")
plt.xlabel("Time (day)")
plt.ylabel("Flowrate (bbl/day) ")
plt.grid()
plt.legend()
```

**Harmonic Decline**

The harmonic decline curve is a variation of the exponential and hyperbolic decline models, where the rate of decline decreases in a manner similar to the hyperbolic model but in a more gradual manner.

```python
#Calculating the flowrate for harmonic decline
b=1
q_h=q_i/(1+d*t)

#Cumulative production
q_h_cum=np.log(q_i/q_h)*(q_i/d)
```

```python
#Creating datframe
df=pd.DataFrame({"time":t,"q":q_h,"q_cum":q_h_cum})
df.head()
```

```python
#Plotting the data
plt.figure(figsize=(15,5))
plt.plot(df["time"],df["q"],lw=4,label="Flowrate vs time",color="green")
plt.title("Plot For Harmonic Decline")
plt.xlabel("Time (day)")
plt.ylabel("Flowrate (bbl/day)")
plt.grid()
plt.legend()
plt.show()
```

**Hyperbolic Decline** -----> **0<b<1**

Hyperbolic decline allows for the rate of decline to decrease over time.

```python
#Take b as 0.4
b=0.4
```

```python
#Calculating flow rate for hyperbolic decline
q_hyp=q_i/np.power((b*d*t+1),(1/b))
q_hyp_cum=(q_i**b)*(q_hyp**(1-b)-q_i**(1-b))/((b-1)*d)
```

```python
#Creating Dataframe
df=pd.DataFrame({"time":t,"q":q_hyp,"q_cum":q_hyp_cum})
df.head()
```

```python

#Plotting the data
plt.figure(figsize=(15,5))
plt.plot(df["time"],df["q"],lw=4,label="Flowrate vs time")
plt.title("Plot for Hyperbolic Decline")
plt.xlabel("Time (day)")
plt.ylabel("Flowrate (bbl/day)")
plt.grid()
plt.legend()
plt.show()
```

```python
#Plotting data
plt.style.use("classic")
fig,ax=plt.subplots(nrows=3,ncols=1)
fig.set_figwidth(15)
fig.set_figheight(20)

#First Curve--> Exponential Curve
ax[0].plot(t,q_e,color="red",lw=6)
ax[0].set_title("Exponential Decline Curve")
ax[0].set_xlabel("Time (day)")
ax[0].set_ylabel("Q (bbl/day)")
ax[0].grid()
ax[0].legend()


#Second Curve-->Harmonic Curve
ax[1].plot(t,q_h,color="green",lw=6)
ax[1].set_title("Harmonic Decline Curve")
ax[1].set_xlabel("Time (day)")
ax[1].set_ylabel("Q (bbl/day)")
ax[1].grid()
ax[1].legend()


#Third Curve-->Hyperbolic Curve
ax[2].plot(t,q_hyp,color="blue",lw=6)
ax[2].set_title("Hyperbolic Decline Curve")
ax[2].set_xlabel("Time (day)")
ax[2].set_ylabel("Q (bbl/day)")
ax[2].grid()
ax[2].legend()
```

```python
#Plot on same graph
plt.style.use("classic")
plt.figure(figsize=(20,10))
plt.plot(t,q_e,color="red",label="Exponential Decline Curve",lw=5)
plt.plot(t,q_h,color="green",label="Harmonic Decline Curve",lw=5)
plt.plot(t,q_hyp,"b",label="Hyperbolic Decline Curve",lw=5)
plt.title("Decline Curve")
plt.xlabel("Time (day)")
plt.ylabel("Q (bbl/day)")
plt.legend()
plt.grid()
plt.legend()
```

```python
#Cumulative production curve
plt.style.use("classic")
plt.figure(figsize=(16,12))
plt.plot(t,q_e_cum/100000,color="red",label="Exponential cumulative production Curve",lw=5)
plt.plot(t,q_h_cum/100000,color="green",label="Harmonic cumulative production Curve",lw=5)
plt.plot(t,q_hyp_cum/100000,"b",label="Hyperbolic cumulative productionCurve",lw=5)
plt.title("Cumulative production in stb")
plt.xlabel("Time (day)")
plt.ylabel("Q_cum X 100000 (bbl)")
plt.legend()
plt.grid()
plt.legend()
```
